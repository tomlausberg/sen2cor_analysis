```python
import os
from pathlib import Path
import pandas as pd
```

```python

out_file = Path(r"C:\Users\tlaus\Documents\local\semester_thesis\sen2cor_analysis\outputs\sen2cor_sn7_with_mods_roi_min_errors.txt")

# read output file line by line
with open(out_file, "r") as f:
    lines = f.readlines()

data = []
for line in lines:
    if line.startswith("Running sen2cor for"):
        print(line)
        current_label = line.split("for ")[-1].replace("\n", "")
        print(current_label)
        location, mod = current_label.replace(" ","").split(":")

    if "Application terminated successfully" in line:
        data.append(
            {
                "location": location,
                "mod": mod,
                "success": True
            }
        )

    elif "Application terminated with at least one error" in line:
        data.append(
            {
                "location": location,
                "mod": mod,
                "success": False
            }
        )
df = pd.DataFrame(data)
```

```python
df.sort_values(by=["location", "mod"], inplace=True)
```

```python
# check if all locs are completely successful
for loc in df.location.unique():
    loc_df = df[df.location == loc]
    if not loc_df.success.all():
        print(f"{loc} has not all mods successful")
```
